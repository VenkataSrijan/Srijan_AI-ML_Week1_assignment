```python
# Task 1: Python Basics & Operators

a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

# Arithmetic Operations
print(f"Addition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} * {b} = {a * b}")
print(f"Division: {a} / {b} = {a / b}")
print(f"Modulus: {a} % {b} = {a % b}")
print(f"Power: {a} ** {b} = {a ** b}")

# Comparison Operators
print(f"{a} > {b} : {a > b}")
print(f"{a} < {b} : {a < b}")
print(f"{a} == {b} : {a == b}")
print(f"{a} != {b} : {a != b}")

# Logical Operators
print(f"(a > 0 and b > 0): {a > 0 and b > 0}")
print(f"(a > 0 or b > 0): {a > 0 or b > 0}")
print(f"not(a > b): {not(a > b)}")
```

```python
#task 2
import random
import numpy as np

# Create list of 10 random integers
numbers = [random.randint(1, 100) for _ in range(10)]
print("Original List:", numbers)

# Add and remove elements
numbers.append(50)
numbers.remove(numbers[0])
print("Updated List:", numbers)

# Max, Min, Sort
print("Maximum:", max(numbers))
print("Minimum:", min(numbers))
numbers.sort()
print("Sorted List:", numbers)

# Convert to NumPy array
arr = np.array(numbers)

print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
```

```python
#Task 3
# Student dictionary
student = {
    "name": "Srijan",
    "course": "AI & ML",
    "marks": 82
}

# Assign grade
if student["marks"] >= 85:
    student["grade"] = "A"
elif student["marks"] >= 70:
    student["grade"] = "B"
else:
    student["grade"] = "C"

# Print dictionary
for key, value in student.items():
    print(f"{key}: {value}")

# Sets of AI tools
ai_tools_1 = {"ChatGPT", "TensorFlow", "PyTorch"}
ai_tools_2 = {"PyTorch", "Keras", "Scikit-learn"}

print("Union:", ai_tools_1 | ai_tools_2)
print("Intersection:", ai_tools_1 & ai_tools_2)
```

```python
#Task 4
# Write student details
with open("ai_students.txt", "w") as file:
    file.write("Ravi,80,A\n")
    file.write("Sneha,92,A\n")
    file.write("Aman,65,C\n")
    file.write("Pooja,78,B\n")
    file.write("Kiran,55,C\n")

# Read and display students above 75 marks
print("Students scoring above 75:")
with open("ai_students.txt", "r") as file:
    for line in file:
        name, marks, grade = line.strip().split(",")
        if int(marks) > 75:
            print(f"{name} - {marks} ({grade})")
```
